```python
import json
# Más tarde, puedes cargar este archivo JSON en tu programa Python
with open('epigrams_processed.json', 'r', encoding='utf-8') as f:
    loaded_epigrams = json.load(f)
```

```python
def print_processed_epigrams(processed_epigrams):
    for ref, content in processed_epigrams.items():
        print(f"{ref}:\n")
        for line_info in content['processed_lines']:
            line_number = line_info['line']
            words = line_info['words']
            # Une las palabras y signos de puntuación para formar la línea del epigrama
            reconstructed_line = ''.join(words)
            # Agrega espacios entre palabras y signos de puntuación para una mejor legibilidad
            reconstructed_line = ''.join([word if word in [',', '.', ':', ';', '!', '?', '’'] else f' {word}' for word in words]).strip()
            if line_number == 5:
                print(f"{line_number}\t{reconstructed_line}")
            else:
                print(f"\t{reconstructed_line}")
        print("\n")  # Espacio extra entre epigramas para mejor lectura

# Llamada a la función para imprimir los epigramas procesados
print_processed_epigrams(loaded_epigrams)
```

```python
worders = []
for i in loaded_epigrams:
    lines = loaded_epigrams[i]["processed_lines"]
    rango = len(lines) - 1
    for line_info in loaded_epigrams[i]["processed_lines"][0:int(rango)]:
        words = line_info["words"]
        worders.append(words)
```

```python
from cltk import NLP

# Initialize the NLP engine for Latin
cltk_nlp = NLP(language='grc', suppress_banner=True)

# Example Latin sentence
latin_text = 'τοῦτο βίος , τοῦτ αὐτό : τρυφὴ βίος . ἔρρετ ἀνῖαι :'

# Process the text
doc = cltk_nlp.analyze(text=latin_text)

# Access morphological information for each word
for word in doc.words:
    print(f'Original: {word.string}, Lemma: {word.lemma}, Features: {word.features}')
```

```python
def print_processed_epigrams_with_analysis(processed_epigrams):
    for ref, content in processed_epigrams.items():
        print(f"{ref}:\n")
        for line_info in content['processed_lines']:
            line_number = line_info['line']
            words = line_info['words']
            # Une las palabras y signos de puntuación para formar la línea del epigrama
            reconstructed_line = ' '.join([word if word in [',', '.', ':', ';', '!', '?', '’'] else word for word in words]).strip()
            
            # Analiza cada palabra con CLTK
            doc = cltk_nlp.analyze(text=' '.join(words))
            analysis_output = []
            for word in doc.words:
                lemma = word.lemma
                features = word.features
                analysis_output.append(f"Original: {word.string}, Lemma: {lemma}, Features: {features}")
            
            # Imprime la línea reconstruida y su análisis morfológico
            if line_number == 5:
                print(f"{line_number}\t{reconstructed_line}")
            else:
                print(f"\t{reconstructed_line}")
            
            # Imprime el análisis morfológico de cada palabra
            for analysis in analysis_output:
                print(f"\t\t{analysis}")
        print("\n")  # Espacio extra entre epigramas para mejor lectura

# Llamada a la función para imprimir los epigramas procesados con análisis morfológico
print_processed_epigrams_with_analysis(loaded_epigrams)
```
